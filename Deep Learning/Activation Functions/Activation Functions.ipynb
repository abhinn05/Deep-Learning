```python
import math

def sigmoid(x):
  return 1/(1+ math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(50)
```

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x) / (math.exp(x) + math.exp(-x)))
```

```python
tanh(56)
```

```python
tanh(-56)
```

```python
def relu(x):
  return max(0,x)
```

```python
relu(-56)
```

```python
relu(4)
```

```python
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(45)
```

```python
leaky_relu(-56)
```
